```python
ufos["month"] = ufos["date_time"].dt.month
```

```python
ufos["month"].value_counts().sort_index()
```

```python
ufos.month.value_counts().sort_index().plot(kind="bar", title="UFO Sightings By Month");
```

```python
sightings = ufos.month.value_counts().sort_index()
```

```python
sightings.plot(kind="bar",
               title="UFO sightings By Month",
               xlabel="Month");
```

```python
sightings.plot(kind="bar",
               title="UFO sightings By Month",
               xlabel="Month",
               ylabel="Num Sightings",
               fontsize=12);
```

```python
sightings.plot(kind="bar",
               title="UFO sightings By Month",
               fontsize=12)
plt.xlabel("Month", fontsize="20", color="olive")
plt.ylabel("Nums Sightings", fontsize="24", color="red");
```

```python
month_dict = ({1: "Jan", 2: "Feb", 3:"Mar", 4: "Apr", 5: "May", 6: "Jun", 7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"})
sightings.rename(month_dict)
```

```python
sightings.rename(month_dict).plot(kind="pie");
```

```python
sightings.rename(month_dict).plot(kind="bar",
               title="UFO sightings By Month",
               fontsize=16)
plt.xlabel("Month", fontsize="20", color="olive")
plt.ylabel("Nums Sightings", fontsize="24", color="red");
```

```python
salaries = pd.read_csv("data/Salaries.csv", low_memory=False)
```

```python
salaries["BasePay"] = pd.to_numeric(salaries["BasePay"], errors="coerce").fillna(0)
salaries["OvertimaPay"] = pd.to_numeric(salaries["OvertimePay"], errors="coerce").fillna(0)
salaries["OtherPay"] = pd.to_numeric(salaries["OtherPay"], errors="coerce").fillna(0)
```

```python
salaries
```

```python
df = salaries[["EmployeeName", "OvertimePay", "BasePay", "OtherPay"]]
```

```python
df
```

```python
df.set_index("EmployeeName", inplace=True)
```

```python
df
```

```python
df.head(10).plot(kind="bar", stacked=True);
```

```python
df["BasePay"].sort_values(ascending=False).head(10).plot(kind="bar");
```

```python
df["BasePay"].sort_values(ascending=False).head(10).plot.bar(color="red", title="Highest Paid SF Employees");
```

```python
df["BasePay"].sort_values(ascending=False).head().plot.barh(color="red", title="Highest Paid SF Employees");
```

```python
df.sort_values("OvertimePay", ascending=False).head().plot(kind="bar");
```

```python
df.sort_values("OvertimePay", ascending=False).head().plot.barh(stacked=True);
```

```python
df.sort_values("OvertimePay", ascending=False).head().plot(kind="barh", stacked=True)
plt.xlabel("Dollars", fontsize=15, color="purple");
```
