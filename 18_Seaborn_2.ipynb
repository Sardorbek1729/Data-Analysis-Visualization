# Countplot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins
```

```python
sns.countplot(data=penguins, x="species");
```

```python
sns.countplot(data=penguins, x="species", hue="sex");
```

```python
titanic = pd.read_csv("data/titanic.csv")
```

```python
titanic
```

```python
sns.countplot(data=titanic, x="pclass", hue="sex");
```

# Stripplot & Swarmplot

```python
taxis = sns.load_dataset("taxis")
```

```python
taxis
```

```python
sns.scatterplot(data=taxis, x="pickup_borough", y="distance");
```

```python
sns.stripplot(data=taxis, x="pickup_borough", y="distance");
```

```python
taxis_sample = taxis.nlargest(600, "total")
```

```python
plt.figure(figsize=(12, 5))
sns.swarmplot(data=taxis_sample, x="pickup_borough", y="distance");
```

```python
sns.stripplot(data=taxis_sample, x="pickup_borough", y="distance");
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
plt.figure(figsize=(12, 5))
sns.swarmplot(data=titanic, x="pclass", y="age");
```

```python
plt.figure(figsize=(12, 5))
sns.swarmplot(data=titanic, x="pclass", y="age", hue="sex");
```

# Catplot

```python
sns.catplot(data=titanic, x="sex", y="survived", kind="bar");
```

```python
sns.catplot(data=titanic, x="sex", y="survived", kind="bar", col="pclass");
```

```python
sns.catplot(data=titanic, x="pclass", y="survived", kind="bar", hue="sex");
```

```python
sns.catplot(data=taxis, x="pickup_borough", y="total", kind="point", col="color");
```

```python
sns.catplot(data=taxis, x="pickup_borough", y="total", kind="violin", hue="color", split=True);
```

```python
sns.catplot(data=titanic, x="who", y="survived", kind="bar", col="pclass", ci=None, row="embarked");
```
