# Bar Plots

```python
plants = ["Spinach", "Turnip", "Rhubarb", "Brocolli", "Kale"]
died = [10,25,5,30,21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, germinated, width=0.9, color="orchid")
plt.title("Germination Rate")
plt.ylabel("Number of Seedlings");
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died, width=0.7, color="red")
plt.title("Failure Rates")
plt.ylabel("Number of Dead Seedlings");
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died)
plt.bar(plants, germinated, width=0.4, align="edge");
```

```python
plt.figure(figsize=(6, 6))
plt.bar(plants, died, bottom=germinated,label="Failed Seedlings")
plt.bar(plants, germinated, label="Germinated Seedlings")
plt.legend(shadow=True, facecolor="white", frameon=True)
plt.title("GreenHouse Seedlings")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.barh(plants, died, color="red");
```

```python
import pandas as pd
titanic = pd.read_csv("data/titanic.csv")
titanic["age"] = pd.to_numeric(titanic["age"], errors="coerce")
```

```python
plt.hist(titanic.age);
```

```python
first_class = titanic[titanic["pclass"] == 1]["age"]
plt.hist(first_class);
```

```python
first_class = titanic[titanic["pclass"] == 1]["age"]
second_class = titanic[titanic["pclass"] == 2]["age"]
third_class = titanic[titanic["pclass"] == 3]["age"]
plt.hist(first_class, label="1st Class")
plt.hist(second_class, label="2nd Class")
plt.hist(third_class, label="3rd Class")
plt.title("Titanic Passenger Ages")
plt.xlabel("Passenger Age")
plt.legend();
```

# Subplots

```python
nums = np.arange(5)
plt.figure(figsize=(10, 4))
plt.suptitle("Out First Subplot", fontsize=18)

plt.subplot(1,3,1)
plt.plot(nums, nums,c="olive")
plt.title("X")

plt.subplot(1,3,2)
plt.plot(nums, nums**2, c="blue")
plt.title("X Squared")

plt.subplot(1,3,3)
plt.plot(nums, nums**3, c="red")
plt.title("X Cubed")
plt.tight_layout();
```

```python
nums = np.arange(5)
plt.figure(figsize=(10, 9))
plt.suptitle("Out First Subplot", fontsize=18)

plt.subplot(3,1,1)
plt.plot(nums, nums,c="olive")
plt.title("X")

plt.subplot(3,1,2)
plt.plot(nums, nums**2, c="blue")
plt.title("X Squared")

plt.subplot(3,1,3)
plt.plot(nums, nums**3, c="red")
plt.title("X Cubed")
plt.tight_layout();
```
